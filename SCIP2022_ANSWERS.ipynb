Because of this bug, when I use colab, in order to highlight specific residues,  I use multiple atom representations. For example, in the code below. We use the licorice representation to show the helices and the beta-sheet. 

```python
view = nv.show_pdbid("1qys")
view.clear_representations()
view.add_licorice('1-26 or 45-56 or 76-94', color='blue')
view.add_licorice('27-44 or 57-75', color="red")
view
```

## 2. What did you notice about the residue numbers for alpha-helices and the strands?

What I wanted you to realize was that in the case of **alpha-helices** all the numbers are contiguous. The first helix goes from residue 27 to residue 44, and the second helix from residue 57 to 75. 

This is not the case for the **beta-sheets**, notice that the sheet is formed by 5 strands, grouped in 3 blocks: 1 to 26, 45 to 56 or 76 to 94. 




## 3. What is the difference between alpha-helices and strands?

What I want you to notice is that alpha-helices are formed by residues interacting with their closests neighbors in sequence. They are formed by interacting with residue +- 4 residues away.  

In the case of a beta-sheet, they are formed between residues that are not nearby. In this structure we can see that there could even be a helix in between! 

I wanted to highlight the difficulty of predicting the final structure a protein will have, based only on its sequence. Since you cannot easily see with which 'residue X' a 'residue Y' could be interacting.
