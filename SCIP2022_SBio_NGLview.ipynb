# How do we get structures?  

There are experimental methods that can allow us to determine the position of the atoms in a molecule. Some of these methods are: Nuclear Magnetic Resonance Spectroscopy, X-ray diffraction, and Cryo-Electron Microscopy. 

The cool thing about structural biology, is that when scientists 'solve' a structure, they share it with everyone! 

The major database of biomolecules is the [Protein Data Bank](https://www.rcsb.org/). To this day the have more than 190,000 biological macromolecular structures.

# Ok, let's get some structures

Let's start with [Top7](https://https://en.wikipedia.org/wiki/Top7). This is a 93-residue $\alpha$/$\beta$ designed protein. It was designed by humans using structure prediction algorithms. You can read more about it here: https://pubmed.ncbi.nlm.nih.gov/14631033/

The PDB ID of Top7 is 1QYS. How do we know this? 

1. You can read the paper. They should mention what is the PDB ID of any protein they report.
2. Type key-words on the search bar of PDB, scroll until you find the one you are looking for. Each structure should mention in which paper they were introduced. As well as more information about the structure. 

We are going to use the function 'show_pdbid' to load the structure, directly from the Protein Data Bank. 

```python
view = nv.show_pdbid("1qys")
```

```python
view.clear_representations()
view
```

```python
view.add_licorice('1-10')
```

```python
view.add_licorice('protein')
```

```python
view.add_cartoon('1-10 or 30-50', color='red')
```

```python
view.add_spacefill('protein')
```

```python
# Save image to your computer 

view.download_image("Top7_SCIP.png")
```
